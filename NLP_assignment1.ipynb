```python
import nltk
from nltk.corpus import gutenberg

nltk.download('gutenberg')
gutenberg.fileids()
sense_raw = gutenberg.raw("austen-sense.txt")
```

```python
from nltk import Text
tokens = nltk.word_tokenize(sense_raw)   # tokenize it
sense = nltk.Text(tokens)           # turn text into a NLTK Text object
```


#### 1. Plot the lexical dispersion plot as Figure 1.2 in chapter 1 for the following names </br>


> à Elinor, Marianne, Colonel, Edward, Lucy



```python
sense.dispersion_plot(["à Elinor", "Marianne", "Colonel", "Edward", "Lucy"])
```

#### 4. Compute the lexical diversity of this novel


```python
def lexical_diversity(text):
  
  return len(set(text)) / len(text)
```

```python
lexical_diversity(sense)
```

#### 5. Find 50 most frequently used words and plot the cumulative frequency plot as Figure 3.2 in chapter 1



```python
from nltk.book import FreqDist

fdist1 = FreqDist(sense)
```

```python
fdist1.plot(50, cumulative=True)
```

```python
[w for w in set(sense) if w.endswith('ed')]
```
